# Investigating the Impact of Maternal Mental Health on Infant Sleep Patterns

**Problem Statemnt**

The objective of this study is to investigate the impact of maternal mental health on infant sleep patterns and birth trauma. Specifically, we aim to examine the relationship between maternal anxiety, depression, and birth trauma with infant sleep duration and frequency of awakenings. By assessing the Hospital Anxiety and Depression Scale (HADS), Edinburgh Postnatal Depression Scale (EPDS), and City Birth Trauma Scale (CBTS), we will gain insights into the psychological well-being of mothers and its potential influence on their infants' sleep.

**Dataset Description**

This dataset presents valuable insights into the relationship between maternal mental health and infant sleep patterns. With a focus on postpartum depression, anxiety, and childbirth-related posttraumatic stress disorder (CB-PTSD), the study encompasses data from 410 mothers and their infants, aged between 3 to 12 months old. Through standardized questionnaires, comprehensive information was gathered on the mothers' psychological well-being, sociodemographic factors (including age, marital status, and educational level), and important infant characteristics (such as age and temperament). Additionally, the dataset captures specific sleeping habits, allowing for a profound exploration of how maternal mental health influences the quality of an infant's sleep.

**Variables of Interest**

In our project, we are interested in understanding the factors that affect the sleep duration and awakening number of infants. To investigate this, we have selected the following variables:

EPDS: This variable represents the level of depressive symptoms in mothers. We hypothesize that higher levels of depressive symptoms may be associated with shorter sleep duration and more frequent awakenings in infants.

HADS_Anxiety: This variable represents the level of anxiety symptoms in mothers. We suspect that higher levels of anxiety may also contribute to shorter sleep duration and increased awakenings in infants.

CBTS: This variable refers to the method of falling asleep for infants, such as whether they are breastfed, rocked, or soothed in other ways. We anticipate that different methods of falling asleep may influence sleep duration and awakening patterns.

Sleep_night_duration_bb1: This is the dependent variable that represents the duration of sleep during the night for infants. We aim to explore how the other variables in our analysis relate to variations in sleep duration.

Gestationnal_age: This variable denotes the gestational age of the infants, indicating how many weeks they were in the womb before birth. We believe that gestational age might impact sleep patterns, with premature infants potentially experiencing different sleep behaviors compared to full-term infants.

Age: This variable represents the age of the infants. We assume that sleep patterns may change as infants grow older, so we want to examine how age influences sleep duration and awakening number.

night_awakening_number_bb1: This is another dependent variable that indicates the number of awakenings during the night for infants. We aim to understand the factors that contribute to variations in the frequency of night awakenings.

```python
def explore_categorical_var(df, x):
    """Returns the numerical and graphical summary of categorical variables, including percentage of each label"""
    value_counts = df[x].value_counts()
    print(value_counts, "\n")
    print("===============================")
    print("Percentage \n")
    total_count = value_counts.sum()
    percentages = (value_counts / total_count) * 100
    print(percentages, "\n")
    print("===============================")
    print(f"Missing values: {df[x].isnull().sum()}")

    # Sort value_counts and percentages in descending order
    value_counts = value_counts.sort_values(ascending=False)
    percentages = percentages.loc[value_counts.index]

    plt.figure()
    ax = sns.countplot(data=df, x=x, order=value_counts.index)
    plt.title(x)

    for i, v in enumerate(percentages):
        ax.text(i, v, f"{v:.2f}%", ha='center', va='bottom', fontweight='bold')

    plt.show()
```

**Principal Component Analysis (PCA)**

In this analysis, we will be using Principal Component Analysis (PCA) for Exploratory Data Analysis (EDA) to gain insights into the relationships between variables and identify important factors that contribute to the variation in our data. 

Through PCA, we aim to uncover the underlying structure in the data and determine the key factors that explain the most variation. By visualizing the data in a reduced dimensional space, we will be able to identify patterns, assess the importance of each variable, and potentially identify any multicollinearity or redundancy among the variables.

The ultimate goal of this analysis is to use the insights gained from PCA to inform my subsequent regression modeling. By understanding the relationships and importance of different variables, we can make informed decisions about which variables to include in the regression model, potentially reducing dimensionality and improving interpretability.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select the variables for PCA
variables = ['EPDS', 'HADS_Anxiety', 'CBTS', 'Gestationnal_age', 'Age']
X = df[variables]
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# Initialize PCA with the desired number of components
pca = PCA(n_components=len(X.columns))

# Fit PCA on your data
pca.fit_transform(scaled_data)

# Get the explained variance ratio for each component
explained_variances = pca.explained_variance_ratio_*100

# Print the explained variance for each component
for i, variance in enumerate(explained_variances):
    print(f"PCA {i+1}: {variance:.4f}")
```

```python
# Fit PCA and obtain explained variance ratios
pca = PCA()
pca.fit(X)
explained_variance = pca.explained_variance_ratio_

# Create scree plot
plt.figure(figsize=(8, 6))
plt.plot(np.arange(1, len(explained_variance) + 1), np.cumsum(explained_variance), marker='o', linestyle='-', color='b')
plt.xlabel('Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.show()
```

```python
def create_pca_scatter_plot(data, dependent_variable):
    # Perform PCA
    pca = PCA(n_components=3)
    pca_components = pca.fit_transform(data[variables])

    # Extract the PCA 1, PCA 2, and PCA 3 values
    pca1 = pca_components[:, 0]
    pca2 = pca_components[:, 1]
    pca3 = pca_components[:, 2]

    # Color code based on the dependent variable
    color_code = data[dependent_variable]

    # Create the scatter plot
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    scatter = ax.scatter(pca1, pca2, pca3, c=color_code, cmap='viridis')
    ax.set_xlabel('PCA 1')
    ax.set_ylabel('PCA 2')
    ax.set_zlabel('PCA 3')
    ax.set_title('PCA Scatter Plot')
    fig.colorbar(scatter, label=dependent_variable)
    plt.show()
```

**Factor Analysis**

The factor analysis in this project aims to uncover the underlying factors related to maternal mental health and birth trauma, providing valuable insights into the structure and dimensions of these constructs. This analysis will facilitate a more nuanced understanding of the factors influencing maternal mental health and their implications for infant sleep patterns and birth trauma experiences.

The eigenvalues approach allows us to identify the significant factors in a factor analysis by examining the magnitude of the eigenvalues. It provides a quantitative basis for selecting the number of factors to retain and helps ensure that the retained factors explain a substantial amount of variance in the data

Eigenvalues represent the amount of variance explained by each factor. Factors with eigenvalues greater than 1.0 indicate that they explain a substantial amount of variance in the data. Therefore, we select the number of factors as three based on the eigenvalues

```python
# Select the variables of interest for factor analysis
variables = ['EPDS', 'HADS_Anxiety', 'CBTS', 'Sleep_night_duration_bb1', 'Gestationnal_age', 'Age', 'night_awakening_number_bb1']

# Create a subset DataFrame with the selected variables
df_subset = df[variables]
```

```python
from factor_analyzer import FactorAnalyzer

# Perform factor analysis to obtain eigenvalues
efa = FactorAnalyzer(n_factors=len(df_subset.columns), method='ml', rotation='varimax')
efa.fit(df_subset)

# Get the eigenvalues
eigenvalues = efa.get_eigenvalues()

# Print the eigenvalues
print("Eigenvalues:")
print(eigenvalues)
```

Based on the eigenvalues obtained, we can select the number of factors for the analysis. In this case, the eigenvalues suggest that the first three factors have eigenvalues greater than 1.0, while the remaining factors have eigenvalues less than 1.0.

```python
# Check for missing values and handle them if necessary
df_subset = df_subset.dropna()

# Performing factor analysis with the selected number of factors
num_factors = 3  
efa = FactorAnalyzer(n_factors=3, method='ml', rotation='varimax')
efa.fit(df_subset)

# Extract factor loadings and communalities
loadings = efa.loadings_
communalities = efa.get_communalities()

# Print the factor loadings and communalities
print("Factor Loadings:")
print(loadings)
print("\nCommunalities:")
print(communalities)
```

**Intepretation of Factor Analysis**

- The first factor has high loadings on Edinburgh Postnatal Depression Scale (EPDS), HADS_Anxiety, and CBTS variables. This indicates a strong association between these variables and the first factor. It suggests that maternal depressive symptoms, anxiety, and birth trauma are closely related and contribute to a common underlying factor.

- The second factor has moderate loadings on Sleep_night_duration_bb1, Gestationnal_age, and Age variables. This suggests that these variables share a common underlying factor related to the duration of sleep, gestational age, and age of the infants.

- The third factor has low loadings on night_awakening_number_bb1 and Age variables. This suggests a weak association between these variables and the third factor.

The interpretation suggests that there are distinct factors related to maternal mental health, infant sleep duration and age, and infant night awakening. These findings contribute to our understanding of the factors influencing sleeping patterns in infants

In all, the factor analysis results suggest that there are underlying factors that explain the relationships between the variables. Specifically, factors related to maternal mental health, the method of falling asleep for infants, sleep duration and age/gestational age of infants, and night awakening patterns have been identified. These factors capture the shared variance among the variables, allowing us to understand how they relate to each other and contribute to infants' sleep duration and awakening patterns.

**Post Hoc Analysis**

After conducting the ANOVA test to assess the effect of the how infant fall asleep on the sleep duration, we performed a post hoc analysis using Tukey's Honestly Significant Difference (HSD) test. The aim of this analysis was to determine which categories of the variable method of falling asleep are significantly different from each other in terms of their impact on the sleep duration. Tukey's HSD test allows us to compare all possible pairs of categories and identify significant differences based on mean differences, standard errors, confidence intervals, and p-values. Perfroming post hoc analysis, we gain insights into which specific variables within 'how_falling_asleep' are associated with distinct sleep duration patterns."

```python
import statsmodels.stats.multicomp as mc

# Perform ANOVA
model = ols('Sleep_night_duration_bb1 ~ C(how_falling_asleep_bb1)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Perform Tukey's HSD test
posthoc = mc.MultiComparison(df['Sleep_night_duration_bb1'], df['how_falling_asleep_bb1'])
tukey_results = posthoc.tukeyhsd()

print(tukey_results)
```

The method of falling asleep, such as being fed, being rocked, being held, or being alone in the crib, does not significantly affect sleep duration in most cases. However, there is a significant difference in sleep duration when comparing infants who fall asleep alone in the crib to those who fall asleep while being fed, being rocked, or with parental presence in the crib.
